```python
#bt3
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

hours = np.arange(1, 11)                              
noise = np.random.normal(0, 0.5, size=10)              
score = hours * 1.2 + noise                           

df = pd.DataFrame({
    "hours": hours,
    "score": score
})

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="hours", y="score", s=80)

plt.title("Mối quan hệ giữa số giờ học và điểm thi")
plt.xlabel("Số giờ học")
plt.ylabel("Điểm số")

plt.grid(True)
plt.show()

```

```python

#bt4
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "class": np.random.choice(["A", "B", "C"], size=60),
    "score": np.random.randint(40, 100, size=60)
})

plt.figure(figsize=(8, 6))
sns.barplot(data=df, x="class", y="score", estimator=np.mean, errorbar=None)

plt.title("Điểm trung bình của từng lớp")
plt.xlabel("Lớp")
plt.ylabel("Điểm trung bình")

plt.show()

df.groupby("class")["score"].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "major": np.random.choice(["CNTT", "Kinh Tế", "Ngôn Ngữ"], size=150),
    "score": np.random.randint(40, 100, size=150)
})

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="major", y="score")
plt.title("Boxplot điểm số theo nhóm ngành")
plt.xlabel("Ngành học")
plt.ylabel("Điểm số")
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x="major", y="score")
plt.title("Violinplot phân phối điểm theo nhóm ngành")
plt.xlabel("Ngành học")
plt.ylabel("Điểm số")
plt.show()

df.groupby("major")["score"].median()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(42)  

day = np.arange(1, 30 + 1)                     
noise = np.random.normal(0, 10, size=30)      
sales = 200 + day * 3 + noise                

df = pd.DataFrame({
    "day": day,
    "sales": sales
})

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="day", y="sales", marker="o")
plt.title("Lineplot - Doanh thu theo ngày")
plt.xlabel("Ngày")
plt.ylabel("Doanh thu")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
sns.regplot(data=df, x="day", y="sales", scatter_kws={"s": 40})
plt.title("Regplot - Xu hướng doanh thu theo thời gian")
plt.xlabel("Ngày")
plt.ylabel("Doanh thu")
plt.grid(True)
plt.show()

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(42)

month = np.arange(1, 13)
category = np.random.choice(["Shoes", "Clothes", "Bags"], size=12)
revenue = 100 + month * 8 + np.random.normal(0, 20, size=12)

df = pd.DataFrame({
    "month": month,
    "category": category,
    "revenue": revenue
})

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="month", y="revenue", hue="category", marker="o")
plt.title("Doanh thu theo tháng cho từng nhóm sản phẩm")
plt.xlabel("Tháng")
plt.ylabel("Doanh thu")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x="category", y="revenue")
plt.title("Phân phối doanh thu theo nhóm sản phẩm")
plt.xlabel("Nhóm sản phẩm")
plt.ylabel("Doanh thu")
plt.show()

plt.figure(figsize=(10, 5))
sns.regplot(data=df, x="month", y="revenue", scatter_kws={"s": 40})
plt.title("Đường hồi quy xu hướng doanh thu theo thời gian (tổng)")
plt.xlabel("Tháng")
plt.ylabel("Doanh thu")
plt.grid(True)
plt.show()

df.groupby("category")["revenue"].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(42)

month = np.arange(1, 13) 
category = np.random.choice(["Shoes", "Clothes", "Bags"], size=12)
revenue = 200 + month * 10 + np.random.normal(0, 30, size=12)

df = pd.DataFrame({
    "month": month,
    "category": category,
    "revenue": revenue
})

plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x="month", y="revenue", hue="category", marker="o")
plt.title("Doanh thu theo tháng của từng nhóm sản phẩm")
plt.xlabel("Tháng")
plt.ylabel("Doanh thu")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
sns.violinplot(data=df, x="category", y="revenue")
plt.title("Phân phối doanh thu theo nhóm sản phẩm")
plt.xlabel("Nhóm sản phẩm")
plt.ylabel("Doanh thu")
plt.show()

plt.figure(figsize=(10, 5))
sns.regplot(data=df, x="month", y="revenue", scatter_kws={"s": 40})
plt.title("Xu hướng doanh thu theo thời gian (Regplot)")
plt.xlabel("Tháng")
plt.ylabel("Doanh thu")
plt.grid(True)
plt.show()

df
```
