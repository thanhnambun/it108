```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)

y = x**2

plt.plot(x, y)

plt.title("Biểu đồ đầu tay của Nam")

plt.xlabel("Trục x")
plt.ylabel("Trục y")

plt.grid(True, linestyle="--")

# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt

months = list(range(1, 13))
rainFalls = [3.1, 2.8, 3.5, 3.9, 4.2, 4.0, 3.8, 3.6, 3.7, 3.9, 4.1, 3.4]

temperatures = [30.5, 32.0, 35.5, 38.0, 40.5, 42.0, 41.5, 40.0, 38.5, 35.0, 32.5, 30.0]

fig,ax = plt.subplots()
ax.plot(months, temperatures, color='red', marker='o', label='Temperature (°C)')
ax.scatter(months, rainFalls, color='blue', marker='x', label='Rainfall (cm)')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt 

x = range(0, 11)
y = [2*x + 3 for x in x]
plt.plot(x, y, color='blue', linewidth=3)
plt.title("Đường thẳng y = 2x + 3")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 200)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='blue', linestyle='-', label='sin(x)')

plt.plot(x, y2, color='red', linestyle='--', label='cos(x)')

plt.title("So sánh sin(x) và cos(x)")

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt 
x = range(0, 11)
y1 = [2*x + 1 for x in x]
y2 = [-x + 10 for x in x]
plt.plot(x, y1, color='green', label='y = 2x + 1')
plt.plot(x, y2, color='orange', label='y = -x + 10')
plt.title("Biểu đồ hai đường thẳng")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

years = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
population = [92, 93, 94, 95, 96, 97, 98, 99, 100, 101]

plt.plot(years, population, color='blue', linestyle='-', marker='o', label='Dân số')

plt.title("Tăng trưởng dân số Việt Nam 2015–2024")

plt.xlabel("Năm")
plt.ylabel("Dân số (triệu người)")

plt.grid(True)

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

products = ["Áo", "Quần", "Giày", "Túi", "Mũ", "Tất"]
sales = [150, 120, 200, 90, 65, 80]

plt.bar(products, sales, color='green', edgecolor='black')

for i, value in enumerate(sales):
    plt.text(i, value + 1, str(value), ha='center', va='bottom', color='white', fontweight='bold')

plt.title("Doanh số bán hàng tháng 10/2025")

plt.xlabel("Sản phẩm")
plt.ylabel("Doanh số (triệu)")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 200)
y = x**3

plt.plot(x, y, color='red', linewidth=3)

plt.title("Đồ thị hàm bậc ba y = x³")

plt.xlabel("Giá trị x")
plt.ylabel("Giá trị y")

plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
