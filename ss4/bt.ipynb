```python
products = []

def check_ma(ma):
    for p in products:
        if p[0] == ma:
            return True
    return False

while True:
    print("\n===== MENU QUẢN LÝ SẢN PHẨM =====")
    print("1. Thêm sản phẩm mới")
    print("2. Hiển thị danh sách sản phẩm")
    print("3. Tìm kiếm sản phẩm theo mã")
    print("4. Cập nhật thông tin sản phẩm")
    print("5. Xóa sản phẩm theo mã")
    print("0. Thoát chương trình")
    choice = input("Nhập lựa chọn: ")

    if choice == "1":
        ma = input("Nhập mã sản phẩm: ")

        if check_ma(ma):
            print(" Mã sản phẩm đã tồn tại. Vui lòng nhập mã khác.")
            continue

        ten = input("Nhập tên sản phẩm: ")
        so_luong = int(input("Nhập số lượng: "))
        gia = float(input("Nhập giá sản phẩm: "))

        product = (ma, ten, so_luong, gia)
        products.append(product)
        print("✔ Thêm sản phẩm thành công!")

    elif choice == "2":
        if len(products) == 0:
            print("Danh sách sản phẩm trống.")
        else:
            print("\n--- Danh sách sản phẩm ---")
            for p in products:
                print("Mã:", p[0], "- Tên:", p[1], "- SL:", p[2], "- Giá:", p[3])

    elif choice == "3":
        ma = input("Nhập mã sản phẩm cần tìm: ")
        found = False

        for p in products:
            if p[0] == ma:
                print("\nThông tin sản phẩm:")
                print("Mã:", p[0], "- Tên:", p[1], "- SL:", p[2], "- Giá:", p[3])
                found = True
                break
        
        if not found:
            print(" Không tìm thấy sản phẩm.")

    elif choice == "4":
        ma = input("Nhập mã sản phẩm cần cập nhật: ")
        found = False

        for i in range(len(products)):
            if products[i][0] == ma:
                ten = input("Nhập tên mới: ")
                so_luong = int(input("Nhập số lượng mới: "))
                gia = float(input("Nhập giá mới: "))
                products[i] = (ma, ten, so_luong, gia)
                found = True
                print(" Cập nhật thành công!")
                break

        if not found:
            print(" Không tìm thấy sản phẩm.")

    elif choice == "5":
        ma = input("Nhập mã sản phẩm cần xóa: ")
        found = False

        for i in range(len(products)):
            if products[i][0] == ma:
                products.pop(i)
                found = True
                print(" Xóa thành công")
                break

        if not found:
            print(" Không tìm thấy sản phẩm.")

    elif choice == "0":
        print("Đã thoát chương trình.")
        break

    else:
        print("Lỗi")
```
