```python
# bt1
def tinh_hieu(a, b):
    return a - b

def tinh_tich(a, b):
    return a * b

a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))

hieu = tinh_hieu(a, b)
tich = tinh_tich(a, b)

print(f"Hiệu của {a} và {b} là: {hieu}")
print(f"Tích của {a} và {b} là: {tich}")
```

```python
# bt2
def kiem_tra_so_nguyen_to(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
num = int(input("Nhập một số nguyên: "))
if kiem_tra_so_nguyen_to(num):
    print(f"{num} là số nguyên tố.")
else:
    print(f"{num} không phải là số nguyên tố.")
```

```python
# bt3
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
num = int(input("Nhập một số nguyên không âm: "))
if num < 0:
    print("Vui lòng nhập số không âm.")
else:
    print(f"Giai thừa của {num} là: {factorial(num)}")
    
```

```python
#bt4
def circle_area(radius):
    import math
    return math.pi * radius * radius
num = int(input("Nhập bán kính hình tròn: "))
if num < 0:
    print("Bán kính không thể âm.")
else:
    print(f"Diện tích hình tròn với bán kính {num} là: {circle_area(num)}") 
```

```python
#bt5
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

year = int(input("Nhập năm: "))
if is_leap_year(year):
    print(f"{year} là năm nhuận.")
else:
    print(f"{year} không phải là năm nhuận.")
```

```python
def sum_of_digits(n):
    x = 0
    for digit in str(n):  
        x += int(digit)  
    return x

num = input("Nhập số nguyên dương: ")
if not num.isdigit() or int(num) < 0:
    print("Vui lòng nhập số nguyên dương")
else:
    print(f"Tổng các chữ số của {num} là: {sum_of_digits(num)}")
```

```python
def handleReverseTheNumber(n):
    negative = n < 0
    n = abs(n)
    reversed_num = int(str(n)[::-1])
    return -reversed_num if negative else reversed_num

number = int(input("Nhập một số nguyên: "))
reversed_number = handleReverseTheNumber(number)
print(f"Số đảo ngược của {number} là: {reversed_number}")
```

```python
def is_perfect_number(n):
    if n <= 1:
        return False  
    
    total = 0
    for i in range(1, n // 2 + 1): 
        if n % i == 0:
            total += i
    
    return total == n

num = int(input("Nhập số nguyên: "))

if is_perfect_number(num):
    print(f"{num} là số hoàn hảo.")
else:
    print(f"{num} không phải là số hoàn hảo.")
```
