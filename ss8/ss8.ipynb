```python
import numpy as np

a = np.array([[1], [2], [3], [4]])   
print("a =\n", a)
print("shape a:", a.shape)
b = np.array([[10, 20, 30, 40, 50]])  
print("b =\n", b)
print("shape b:", b.shape)
c = a * b   
print("a * b =\n", c)
sqrt_c = np.sqrt(c)
log_c = np.log(c)

print("sqrt(c) =\n", sqrt_c)
print("log(c) =\n", log_c)
print("Min theo hàng:", np.min(c, axis=1))
print("Max theo hàng:", np.max(c, axis=1))
print("Mean theo hàng:", np.mean(c, axis=1))
print("Std theo hàng:", np.std(c, axis=1))
```

```python
import pandas as pd
import numpy as np

data = {
    "name": ["An", "Bình", "Châu", "Dũng"],
    "age": [20, np.nan, 22, 21],
    "score": [8.5, np.nan, 7.0, 9.0],
    "city": ["Hà Nội", "HCM", np.nan, "Hà Nội"]
}

df = pd.DataFrame(data)
print(df)
print(df.isna())
df["age"] = df["age"].fillna(df["age"].mean())
df["score"] = df["score"].fillna(df["score"].median())
df["city"] = df["city"].fillna("Unknown")
df["city"] = df["city"].astype("category")

def classify_score(x):
    if x >= 8:
        return "Good"
    elif x >= 6:
        return "Average"
    else:
        return "Bad"

df["score_level"] = df["score"].apply(classify_score)
age_mean = df["age"].mean()
age_std = df["age"].std()

df["age_normalized"] = (df["age"] - age_mean) / age_std
print(df)
```

```python
# Bài 8
import os
import json
import pandas as pd
import re

folder = "logs"
records = []

pattern = re.compile(r"log_(\d{4}-\d{2}-\d{2})\.json")

for filename in os.listdir(folder):
    match = pattern.match(filename)
    if not match:
        continue
    date = pd.to_datetime(match.group(1))
    
    filepath = os.path.join(folder, filename)
    with open(filepath, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    records.append({
        "date": date,
        "user": data.get("user"),
        "action": data.get("action"),
        "duration": data.get("duration", 0)
    })

df = pd.DataFrame(records)
report = df.groupby("date")["duration"].sum().reset_index()

report.to_csv("daily_report.csv", index=False)

print(report)
```
