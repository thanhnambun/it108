```python
# Bài 2
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]

df = pd.DataFrame({'name': name, 'age': age, 'score': score})

high_score = df.loc[df["score"] >= 8]
age_column = df.iloc[:, 1]
filtered = df.query("score > 7 and age < 23")

print("DataFrame filtered by loc[]:")
print(high_score)

print("DataFrame filtered by iloc[]:")
print(age_column)

print("DataFrame filtered by query():")
print(filtered)

print("\nDataFrame describe:")
print(df.describe())
```

```python
# Bài 3
import pandas as pd

name = ['Nam', 'Bình', 'Châu', 'Dũng']
age = [20, None, 22, 20]
score = [8.5, None, 6.5, 9.0]
df = pd.DataFrame({'name': name, 'age': age, 'score': score})

print("Dữ liệu thiếu kiểm tra bằng isna()")
print(df.isna())

mean_age = df['age'].mean()
print("\nDataFrame sau khi thay thế NaN bằng fillna():")
print(df.fillna(mean_age))

print("\nDataFrame sau khi loại bỏ NaN bằng dropna():")
print(df.dropna())
```

```python
# Bài 4
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]
df = pd.DataFrame({'name': name, 'age': age, 'score': score})

df['age'] = df['age'].apply(lambda x: 'Young' if x <= 20 else "Adult")
print('DataFrame sau khi áp dụng apply():')
print(df)

score_map = {8.5: "Excellent", 7.0: "Good", 6.5: "Average", 9.0: "Excellent"}
df['score_level'] = df['score'].map(score_map)
print("\nDataFrame sau khi map():")
print(df, "\n")

df['score'] = df['score'].astype(int)
print("DataFrame sau khi astype():")
print(df[['name', 'age', 'score']], "\n")

df['name'] = df['name'].replace("Bình", "Bình (Updated)")
print("DataFrame sau khi replace():")
print(df)
```

```python
# Bài 5
import pandas as pd

name = ['An', 'Bình', 'Châu', 'Dũng']
age = [20, 21, 22, 20]
score = [8.5, 7.0, 6.5, 9.0]
df = pd.DataFrame({'name': name, 'age': age, 'score': score})

grouped = (
    df.groupby('age')['score']
      .agg(mean_score='mean', min_score='min', max_score='max')
      .reset_index()
)

print("Groupby age:")
print(grouped)

print('\n', df)
```

```python
# Bài 6
import pandas as pd

df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['An', 'Bình', 'Châu']})
df2 = pd.DataFrame({'id': [1, 2, 3], 'score': [8.5, 7.0, 6.5]})

merged_df = df1.merge(df2)
print("Merged DataFrame:")
print(merged_df)

df_age = pd.DataFrame({'age': [20, 21, 22]})
concatenated_df = pd.concat([merged_df, df_age], axis=1)
print("\nConcatenated DataFrame:")
print(concatenated_df)
```

```python
# Bài 7
import pandas as pd

date = ['2023-01-01', '2023-02-01', '2023-03-01']
name = ['An', 'Bình', 'Châu']

df = pd.DataFrame({'date': date, 'name': name})
df['date'] = pd.to_datetime(df['date'])

year_extracted = df['date'].dt.year
month_extracted = df['date'].dt.month
day_extracted = df['date'].dt.day

df['name'] = df['name'].astype('category')

print(df)
print("\nExtracted Year:")
print(year_extracted)
print("\nExtracted Month:")
print(month_extracted)
print("\nExtracted Day:")
print(day_extracted)
print("\nCategory Type for name:")
print(df['name'])
```

```python
# Bài 8
import os
import json
import pandas as pd
import re

folder = "logs"
records = []

pattern = re.compile(r"log_(\d{4}-\d{2}-\d{2})\.json")

for filename in os.listdir(folder):
    match = pattern.match(filename)
    if not match:
        continue
    date = pd.to_datetime(match.group(1))
    
    filepath = os.path.join(folder, filename)
    with open(filepath, "r", encoding="utf-8") as f:
        data = json.load(f)
    
    records.append({
        "date": date,
        "user": data.get("user"),
        "action": data.get("action"),
        "duration": data.get("duration", 0)
    })

df = pd.DataFrame(records)
report = df.groupby("date")["duration"].sum().reset_index()

report.to_csv("duration_report.csv", index=False)

print(report)
```
