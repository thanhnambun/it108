```python
import math

a = int(input("Nhập số thứ 1: "))
b = int(input("Nhập số thứ 2: "))

array = []

if a > b:
    print("Vui lòng nhập lại số thứ 2 lớn hơn số thứ 1!")
else:
    for i in range(a, b + 1):
        if i < 2:
            continue  

        is_prime = True
        for j in range(2, int(math.sqrt(i)) + 1):
            if i % j == 0:
                is_prime = False
                break

        if is_prime:
            array.append(i)
```
