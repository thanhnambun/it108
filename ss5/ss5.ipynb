```python
# Bài 2
def read_file(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
    return content

def write_file(file_path, data):
    with open(file_path, 'w') as file:
        file.write(data)

file_read = 'students.csv'
file_write = 'high_grade_students.csv'
name = input("Nhập tên: ")
age = int(input("Nhập tuổi"))
grade = float(input("Nhập điểm"))

if grade >= 7:
    write_file(file_write, f"{name},{age},{grade}")
write_file(file_read, f"{name},{age},{grade}")

data = read_file(file_read)
print('students: ', data)
new_data = read_file(file_write)
print('high grade students: ',new_data)

```

```python
# Bài 3
import json

def find_user_with_id(id, file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    for i in data['users']:
        if i['id'] == id:
            return i
        
def write_user_to_file(data, file_path):
    with open(file_path, 'w') as file:
        json.dump(data, file, indent=4)

def read_file(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

file_path = 'user_data.json'
data = find_user_with_id(3, file_path)
print("Người dùng với id là 3: ", data)

name = input("Nhập tên")
age = int(input("Nhập tuổi"))

all_data = read_file(file_path)
new_id = max([u["id"] for u in all_data["users"]], default=0) + 1
new_user = {"id": new_id, "name": name, "age": age}
write_user_to_file(all_data, file_path)

all_data["users"].append(new_user)
write_user_to_file(all_data, file_path)
data = read_file(file_path)
print("Danh sách người dùng sau khi được thêm vào: ", data)
```

```python
# Bài 5

def count_file_word(file_path):
    with open(file_path, 'r') as file:
        content = file.read()
    words = content.split()
    return len(words)

file_path = "sample.txt"
count = count_file_word(file_path)
print(F"Số từ trong file {file_path} là: {count}")
```

```python
# Bài 6
import json
import uuid

def read_inventory(file_path):
    with open(file_path, "r", encoding="utf-8") as file:
        return json.load(file)

def write_inventory(file_path, data):
    with open(file_path, "w", encoding="utf-8") as file:
        json.dump(data, file, indent=4, ensure_ascii=False)

def add_product(file_path, new_product):
    data = read_inventory(file_path)
    if isinstance(data, dict) and "products" in data:
        product_list = data["products"]
    elif isinstance(data, list):
        product_list = data
    else:
        raise ValueError("Cấu trúc JSON không hợp lệ")
    product_list.append(new_product)
    write_inventory(file_path, data)
    return data

file_path = "inventory.json"

name = input("Nhập tên sản phẩm")
quantity = int(input("Nhập số lượng sản phẩm"))
price = float(input("Nhập giá sản phẩm"))

new_item = {
    "product_id": str(uuid.uuid4()),
    "name": name,
    "quantity": quantity,
    "price": price
}

updated_inventory = add_product(file_path, new_item)
print("Đã thêm sản phẩm mới:", updated_inventory)
```

```python
# Bài 8
from datetime import datetime

file_path = "log.txt"

today = datetime.today().date()

with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        timestamp_str = line[:19]
        try:
            timestamp = datetime.strptime(timestamp_str, "%Y-%m-%d %H:%M:%S")
            if timestamp.date() == today:
                print(line.strip())
        except ValueError:
            continue

```
